```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# List of categorical columns to encode
categorical_columns = list(df.select_dtypes("object").columns)

# Apply Label Encoding to each categorical column
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])
    print(f"Encoded {column}:")
    print(df[column])  # Print the encoded column
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from imblearn.over_sampling import SMOTE
```

```python
# Define the target column and features
X = df.drop('Class/ASD', axis=1)  # Features
y = df['Class/ASD']  # Target variable
```

```python
# Step 3: Apply SMOTE to balance the classes
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
# Step 4: Use SelectKBest to select the top K features
select_k_best = SelectKBest(score_func=f_classif, k="all")  # Change k to select specific number of features
X_new = select_k_best.fit_transform(X_res, y_res)
```

```python
selected_columns = X.columns[select_k_best.get_support()]
print("Selected columns based on SelectKBest:", selected_columns)
```

## Xgboost

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Train the XGBoost model
xgb_model = xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')

# Fit the model to the resampled data
xgb_model.fit(X_new, y_res)

# Make predictions on the training set (you can use X_test if available for testing data)
y_train_pred = xgb_model.predict(X_new)

# Calculate the accuracy
train_accuracy = accuracy_score(y_res, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.4f}")

# Confusion Matrix
cm_train = confusion_matrix(y_res, y_train_pred)
print("Training Confusion Matrix:")
print(cm_train)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title('Training Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

## LSTM

```python
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.callbacks import EarlyStopping
from keras.utils import to_categorical

# One-hot encode the labels for multi-class classification
y_res_encoded = to_categorical(y_res, num_classes=3)

# Reshape the data for LSTM
X_new_reshaped = X_new.reshape((X_new.shape[0], X_new.shape[1], 1))

# Define the LSTM model with a deeper architecture
lstm_model = Sequential()

# Add the first LSTM layer
lstm_model.add(LSTM(units=32, activation='relu', input_shape=(X_new_reshaped.shape[1], 1), return_sequences=True))

lstm_model.add(LSTM(units=32, activation='relu', return_sequences=True))

lstm_model.add(LSTM(units=64, activation='relu', return_sequences=True))
lstm_model.add(Dropout(0.2))

# Add the second LSTM layer
lstm_model.add(LSTM(units=128, activation='relu', return_sequences=True))
lstm_model.add(Dropout(0.2))

# Add the third LSTM layer
lstm_model.add(LSTM(units=256, activation='relu'))
lstm_model.add(Dropout(0.02))

# Add the output layer for multi-class classification
lstm_model.add(Dense(3, activation='sigmoid'))  

# Compile the model
lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

lstm_model.fit(X_new_reshaped, y_res_encoded, epochs=50, batch_size=64, verbose=1, validation_split=0.2, callbacks=[early_stopping])

y_train_pred = lstm_model.predict(X_new_reshaped)
y_train_pred = y_train_pred.argmax(axis=1)  

# Calculate the accuracy
train_accuracy = accuracy_score(y_res, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.4f}")

# Confusion Matrix
cm_train = confusion_matrix(y_res, y_train_pred)
print("Training Confusion Matrix:")
print(cm_train)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title('Training Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

## Stacking Classifier

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier

# Define the base models
base_learners = [
    ('ada', AdaBoostClassifier()),
    ('svc', SVC(random_state=42)),
    ('knn', KNeighborsClassifier())
]

# Define the meta-model (final estimator)
meta_model = LogisticRegression(random_state=42)

# Create the StackingClassifier
stacking_model = StackingClassifier(estimators=base_learners, final_estimator=meta_model)

# Fit the stacking model to the resampled data
stacking_model.fit(X_new, y_res)

# Make predictions on the training set (you can use X_test if available for testing data)
y_train_pred = stacking_model.predict(X_new)

# Calculate the accuracy
train_accuracy = accuracy_score(y_res, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.4f}")

# Confusion Matrix
cm_train = confusion_matrix(y_res, y_train_pred)
print("Training Confusion Matrix:")
print(cm_train)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title('Training Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
